```python
!pip install google_ngram_downloader

import nltk
import random
from nltk.lm import Lidstone
from google_ngram_downloader  import readline_google_store
import re


nltk.download('punkt')
```

```python
def process_ngram(ngram):
    tokens = ngram.split()

    words = []

    for token in tokens:

        match = re.search(r"(.*)_(\w+)", token)
        if match:
            word = match.group(1)
            if word:
                words.append(word)
        else:
            words.append(token)

    words = [word.replace('_', '').lower() for word in words]
    words = ' '.join(words)
    words = nltk.word_tokenize(words)
    words = ' '.join(words)
    return words
```

```python
n = 1

v = 10000
```

```python
def fetch_ngrams(n, lang):
    try:
        fname, url, records = next(readline_google_store(ngram_len=n, lang=lang))
        return records
    except:
        print("Error fetching the n-grams.")
        return []

# Let's process only the first 10,000 n-grams for demonstration purposes
MAX_NGRAMS = None

ngram_freqs = []
``
processed = 0

for idx, record in enumerate(fetch_ngrams(n, 'spa')):
    if MAX_NGRAMS is not None and idx > MAX_NGRAMS:
        break
    ngram, year, freq, _ = record
    ngram = process_ngram(ngram)
    ngram_freqs.append((ngram, freq))
    processed += 1
    if processed % 50000 == 0:
        print('Ngrams processed', processed)
```

```python
ngrams, vocab = nltk.lm.preprocessing.padded_everygram_pipeline(n, [ngram for ngram, freq in ngram_freqs for _ in range(freq)])

alpha = 0.01
model = Lidstone(order=n, gamma=alpha, vocabulary=nltk.lm.Vocabulary(vocab, unk_cutoff=1))


model.fit(ngrams)
```

```python
import pickle
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)

with open("model.pkl", "rb") as f:
    loaded_model = pickle.load(f)
```
